### Bài 4: Viết chương trình nhập vào số tuổi của 3 người và cho biết ai là người trẻ nhất và ai là người lớn tuổi nhất

````python
age_person1 = int(input("Please input the age of person #1: "))
age_person2 = int(input("Please input the age of person #2: "))
age_person3 = int(input("Please input the age of person #3: "))

oldest = max(age_person1,age_person2,age_person3)
if oldest == age_person1:
    print("The oldest is person #1")
elif oldest == age_person2:
    print("The oldest is person #2")
else:
    print("The oldest is person #3")

youngest = min(age_person1,age_person2,age_person3)
if youngest == age_person1:
    print("The youngest is person #1")
elif oldest == age_person2:
    print("The youngest is person #2")
else:
    print("The youngest is person #3")
````

### Bài 7: Một công ty vận tải có cách tính chi phí giao hàng như sau:
1. 10kilogram đầu tiên: \$5/kg
2. Từ kilogram thứ 11 đến 20: \$7/kg
3. Từ kilogram thứ 20 trở lên: \$10/kg

Viết chương trình nhập vào số kg cần giao(số nguyên) và in ra chi phí giao hàng tương ứng

````python
# Premise: delivery_weight is positive number

delivery_weight = int(input("Please input delivery weight (kg):  "))
if delivery_weight <= 10:
    delivery_fee = delivery_weight * 5
    print(f"The delivery fee is ${delivery_fee}")
elif delivery_weight <= 20:
    delivery_fee = 10 * 5 + (delivery_weight - 10) * 7
    print(f"The delivery fee is ${delivery_fee}")
else:
    delivery_fee = 10 * 5 + (delivery_weight - 10) * 7 + (delivery_weight - 20) * 10
    print(f"The delivery fee is ${delivery_fee}")
````

Nếu giá trị đơn hàng trên 2 triệu, giảm 10% cho khách, trên 4 triệu, giảm 15%

Tính và in ra tổng giá tiền được giảm

Tính và in ra tổng giá tiền khách cần trả

In ra dòng chữ "Cảm ơn quý khách đã mua hàng"

````python
"""Ví dụ:
Nhập số lượng gà rán khách mua:100
Nhập số lượng hamburger khách mua:100
Nhập số lượng pesi khách mua:15
Tổng giá tiền đơn hàng là: 8.650.000
Tổng tiền được giảm là: 1.297.500
Tổng tiền phải thanh toán: 7.352.500
Cảm ơn quý khách đã mua
"""

chicken = int(input("Please input chicken quantity (pcs): "))
hamburger = int(input("Please input hamburger quantity (pcs): "))
pepsi = int(input("Please input pepsi quantity (pcs): "))

bill_total = chicken * 35000 + hamburger * 50000 + pepsi * 10000

if bill_total > 2000000 and bill_total <= 4000000:
    discount = bill_total * 0.1
    bill_total_discount = bill_total - discount
    print(f"Subtotal: {bill_total}")
    print(f"Discount amount: {discount}")
    print(f"Total amount: {bill_total_discount}")
    print("Thanks for your purchase")
elif bill_total > 4000000:
    discount = bill_total * 0.15
    bill_total_discount = bill_total - discount
    print(f"Subtotal: {bill_total}")
    print(f"Discount amount: {discount}")
    print(f"Total amount: {bill_total_discount}")
    print("Thanks for your purchase")
else:
    print(f"Total amount: {bill_total}")
    print("Thanks for your purchase")
````
